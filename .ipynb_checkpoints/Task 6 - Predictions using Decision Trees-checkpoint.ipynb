# Task 6 - Predictions using Decision Trees

```python
import os
import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

from sklearn.preprocessing import StandardScaler, OrdinalEncoder

from sklearn.pipeline import Pipeline

%matplotlib inline
```

## Loading and Preparing the Data (Iris)

```python
iris = pd.read_csv(os.path.join("Iris.csv"))
iris.head()
```

```python
iris.info()
```

```python
assert iris.isna().sum().sum() == 0
```

```python
iris.drop_duplicates(inplace=True)
```

```python
iris.drop("Id", axis="columns", inplace=True)
```

```python
iris.describe().transpose()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 6), constrained_layout=True)

sns.kdeplot(x="SepalLengthCm", data=iris, hue="Species", shade=True, ax=axes[0, 0]);
sns.kdeplot(x="SepalWidthCm" , data=iris, hue="Species", shade=True, ax=axes[0, 1]);
sns.kdeplot(x="PetalLengthCm", data=iris, hue="Species", shade=True, ax=axes[1, 0]);
sns.kdeplot(x="PetalWidthCm" , data=iris, hue="Species", shade=True, ax=axes[1, 1]);
```

```python
variables = iris.select_dtypes("float").columns
g = sns.PairGrid(data=iris, vars=variables, hue="Species")
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.kdeplot, shade=True)
```

**Key Insights**
From the figures above we can indentify alot of things which are.
 - From the Density Plots above we can easly differentiate between the **Iris-Setosa** and the other two very easly using the knowledege of the **Petal Length** only or the **Petal Width** also alone as there is very clear gap between the **Setosa** and the other two types.

```python
## Let's consider the sepal or the petal as just a rectangle with an area of the (length x width)
## and create two new columns `sepal_area` and `petal_area` and compare them to see the difference 
## between the three types.
```

```python
iris["sepal_area"] = iris["SepalLengthCm"] * iris["SepalWidthCm"];
iris["petal_area"] = iris["PetalLengthCm"] * iris["PetalWidthCm"];

iris.head()
```

```python
plt.rcParams["font.size"] = 13

fig, ax = plt.subplots(1, figsize=(10, 6))
sns.scatterplot(x="sepal_area", y="petal_area", data=iris, hue="Species", ax=ax)
ax.set_xlabel("Sepal Area")
ax.set_ylabel("Petal Area")
ax.set_title("Petal Area vs. Sepal Area");
```

As the figure shows also the **Setosa** is clearly seperated from the other two and also we can easly seperate the other two but with some errors that is quit small.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, fmt=".3f")
plt.xticks(rotation=45);
```

```python
SPECIES = iris["Species"].unique()

iris.loc[:, "Species"] = OrdinalEncoder().fit(iris[["Species"]]).transform(iris[["Species"]]).ravel()
iris["Species"] = iris["Species"].astype(np.int32)

iris.head()
```

## Data Modeling

```python
TEST_SIZE = 0.1
RANDOM_STATE = 42

y = iris["Species"]
X = iris.drop("Species", axis="columns")

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE,
                                                    random_state=RANDOM_STATE)

x_train.shape, x_test.shape
assert x_train.shape[-1] == x_test.shape[-1]
```

### Data Transformation (Scalling)

```python
scaler = StandardScaler()

scaler = scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train[:3]
```

```python
clf_dt = DecisionTreeClassifier()
clf_dt.fit(x_train_scaled, y_train)

print("Decision Tree Score : {:.4f}".format(clf_dt.score(x_test_scaled, y_test)))
```

```python
clf_rf = RandomForestClassifier()
clf_rf.fit(x_train_scaled, y_train)

print("Random Forest Score : {:.4f}".format(clf_rf.score(x_test_scaled, y_test)))
```

```python
clf_ab = AdaBoostClassifier()
clf_ab.fit(x_train_scaled, y_train)

print("Ada Boost Score : {:.4f}".format(clf_ab.score(x_test_scaled, y_test)))
```

```python
dt_pred = clf_dt.predict(x_test_scaled)
rf_pred = clf_rf.predict(x_test_scaled)
ab_pred = clf_ab.predict(x_test_scaled)
```

```python
print("Classification Report For {}: \n".format(type(clf_dt).__name__))
print(classification_report(dt_pred, y_test))

print("\nClassification Report For {}: \n".format(type(clf_rf).__name__))
print(classification_report(rf_pred, y_test))

print("\nClassification Report For {}: \n".format(type(clf_ab).__name__))
print(classification_report(ab_pred, y_test))
```

## Production Pipeline

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", RandomForestClassifier())
])

pipe = pipe.fit(x_train_scaled, y_train)

pipe_pred = pipe.predict(x_test_scaled)

print("\nClassification Report For {}: \n".format(type(pipe).__name__))
print(classification_report(pipe_pred, y_test))
```

### Saving the model using `pickle`

```python
with open("rf_pipe_model.pkl", "wb") as f:
    pickle.dump(pipe, f)
    
!ls
```
